# **Introduction to Convolutional Neural Networks (CNNs)**

A Convolutional Neural Network (ConvNet/CNN) is a deep learning algorithm which mainly deals with image inputs.

Convolutional networks were inspired by biological processes in animal brains. The connectivity pattern between artificial neurons in the CNNs resembles that of the organization of biological neurons in the animal visual cortex (the part of brain that is responsible for vision).

As shown in the diagram below, human brains get a vision from the eyes, which later performs some operations to detect various objects. CNNs also do the same. They try to extract features from the image for image detection.

Now that you know some of the applications of CNN. Let us take a look at the architecture of CNN

A convolutional neural network consists of an input layer, hidden layers and an output layer. The hidden layers of a CNN typically consist of *convolutional layers, pooling layers, fully connected layers*. In a convolutional neural network, the hidden layers include layers that perform ***convolutions***, which we'll discuss in upcoming topics.

We already know that Convolutional Neural Network(CNN) is a type of modified artificial neural network. A CNN consists of an input and an output layer, as well as multiple hidden layers. The hidden layers of a CNN typically consist of *convolutional layers, pooling layers, fully connected layers*. Let us understand what these layers are.

#### **Convolutional Layer**

>Convolutional Layers are the layers that are most significant in CNN models, and hence it's name- Convolutional Neural Networks. These layers perform actions that results in the output having different features, which are more simpler to train the model.

Convolutional Layers have a moving filter(matrix) also called as the weight matrix or Kernel. The filter slides over the input image (convolution operation) to produce a feature map.
Weight matrix/Kernel and Input Image multiply (dot product) and summed to produce the feature map.

#### **Pooling layer**

>Pooling layers are used to progressively reduce the spatial size of the input. The main objective is to reduce the dimensions of input matrix and thus, to reduce the training complexity.

We move a window across a 2D input space, where a certain characteristic like Maximum or Average from that space within that window is the output.
It is also called as downsampling layer because it reduces the number of parameters within the model.

#### **Fully Connected Layer**

>Fully Connected Layers are similar to that of hidden layers in regular neural networks.

In Fully Connected layer all neurons are connected to all neurons of the previous layer. After feature extraction we need to classify the data into various classes, this can be done using a fully connected neural network.
This works similar to that of a Regular Feed forward Neural Network.
The Layer has to flatten and is connected to the output layer, that’s what fully connected does, it flattens. It is also called the dense layer.
