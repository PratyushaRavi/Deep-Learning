#Course Overview

The goal of this course is to give leaners a basic understanding about Neural networks and how they work. Learners will be introduced with the architecture of deep neural network and various tools,libraries and algorithms along with hands on experience starting from how to build a simple neural network to some cool projects like sentiment analysis and object recognition.

#Project Details

During the process of learning we will be doing 5 projects that are:

* HOUSE PRICE PREDICTION

---
We will be building a deep learning model which is going to predict whether the price of a given house is greater or lesser than a certain median value

* TIME SERIES PREDICTION

---
We will be building a model in which given a specific event in a series of equally spaced time period, learn to build a model to predict the next event in it's upcoming future



* SENTIMENTAL ANALYSIS

---
We will be building a model which will classify the polarity of a given text at the document/sentence and predict whether the expressed opinion in a document/ sentence  is positive or negative.


* HAND WRITTEN DIGIT RECOGNITION

---
We will be building a model to classify handwritten digits. The goal is to take an image of a handwritten digit and determine what that digit is. The digits range from zero (0) through nine (9).


* OBJECT RECOGNITION IN A PHOTOGRAPH

---
We will be building a model to identify different classes of object from an image having  10 different classes namely airplanes, cars, birds, cats, deer, dogs, frogs, horses, ships, and trucks.
